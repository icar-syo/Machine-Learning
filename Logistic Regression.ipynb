```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('ex2data1.txt',names=['Exam1','Exam2','Score'])
data.head()
```

```python
fig,ax=plt.subplots()
X=data[data['Score']==0]['Exam1']
Y=data[data['Score']==0]['Exam2']
x=data[data['Score']==1]['Exam1']
y=data[data['Score']==1]['Exam2']
ax.scatter(X,Y,c='r',marker='x',label='y=0')
ax.scatter(x,y,c='b',marker='o',label='y=1')
ax.legend()
ax.set(xlabel='Exam1',ylabel='Exam2')

# 显示label

plt.show()
```
